```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gen_circle():          #generates random centers and radius for circles
    x=np.random.uniform(0,1)
    y=np.random.uniform(0,1)
    r=np.random.uniform(0.02,0.1)
    while ((x+r>1) or (x-r<0) or (y-r<0) or (y+r)>1): #checks if circle exceeds box edges
        x=np.random.uniform(0,1)
        y=np.random.uniform(0,1)
    return (x,y,r)
```

```python
x,y,r=gen_circle()

x_List=[x]
y_List=[y]
r_List=[r]

circles=[]
Area=0

while (Area<0.55): #fills upto area=0.55
    x,y,r=gen_circle()
    t=0
    for a,b,c in zip(x_List,y_List,r_List):
        d=np.sqrt(((x-a)**2)+((y-b)**2))   #measures the distance with nearby circles
        if d<(r+c): #if overlaps, increases t
            t+=1        

    if t==0:  #only adds the circle if there is no overlap
        x_List.append(x)
        y_List.append(y)
        r_List.append(r)
        Area+=(np.pi)*(r**2)

#plotting
fig,ax=plt.subplots()
ax.set_aspect(1)
ax.set_facecolor("grey")
ax.set(xlim=(0,1), ylim=(0,1))

for i in range(len(x_List)):
    circle = plt.Circle((x_List[i], y_List[i]), r_List[i],color='tan')
    ax.add_artist(circle)   
plt.show()
```
